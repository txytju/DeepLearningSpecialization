# 使用 ResNet50 进行 ImageNet 分类

```python
from keras.applications.resnet50 import ResNet50
from keras.preprocessing import image
from keras.applications.resnet50 import preprocess_input, decode_predictions
import numpy as np
```

```python
model = ResNet50(weights='imagenet')
```

```python
img_path = '/notebooks/CNN/week2/KerasTutorial/images/car.jpg'
img = image.load_img(img_path, target_size=(224, 224))
x = image.img_to_array(img)
x = np.expand_dims(x, axis=0)
x = preprocess_input(x)

preds = model.predict(x)
# decode the results into a list of tuples (class, description, probability)
# (one such list for each sample in the batch)
print('Predicted:', decode_predictions(preds, top=3)[0])
# 预测结果时 豪华小轿车，正确
```

# 从VGG16的任意中间层中抽取特征

```python
from keras.applications.vgg16 import VGG16
from keras.preprocessing import image
from keras.applications.vgg16 import preprocess_input
import numpy as np
```

```python
# 可以使用下述方式直接让 Keras 下载预训练模型的权重，也可以一次性下载之后自己手动载入
# 由于让 Keras 下载比较慢，这里选择添加自己手动下载的权重
# base_model = VGG16(weights='imagenet', include_top=True)
base_model = VGG16(weights=None)
```

```python
model = Model(inputs=base_model.input, outputs=base_model.get_layer('block4_pool').output)

img_path = 'images/elephant.jpg'
img = image.load_img(img_path, target_size=(224, 224))
x = image.img_to_array(img)
x = np.expand_dims(x, axis=0)
x = preprocess_input(x)

block4_pool_features_random = model.predict(x)
```

# 将模型的训练结果保存，并重新使用

### 当模型训练完成之后，应该将模型的以及模型的参数保存，以便后续进行 predict 时候使用。下面使用一个小规模的数据对这一流程做一个展示。

```python
from keras.models import load_model
```

```python
X_train_orig, Y_train_orig, X_test_orig, Y_test_orig, classes = load_dataset()

# Normalize image vectors
X_train = X_train_orig/255.
X_test = X_test_orig/255.

# Reshape
Y_train = Y_train_orig.T
Y_test = Y_test_orig.T

print ("number of training examples = " + str(X_train.shape[0]))
print ("number of test examples = " + str(X_test.shape[0]))
print ("X_train shape: " + str(X_train.shape))
print ("Y_train shape: " + str(Y_train.shape))
print ("X_test shape: " + str(X_test.shape))
print ("Y_test shape: " + str(Y_test.shape))
```

```python
def HappyModel(input_shape):
    """
    Model demo.
    """
    X_input = Input(input_shape)
    X = ZeroPadding2D((3, 3))(X_input)
    X = Conv2D(32, (7, 7), strides = (1, 1), name = 'conv0')(X)
    X = BatchNormalization(axis = 3, name = 'bn0')(X)
    X = Activation('relu')(X)
    X = MaxPooling2D((2, 2), name='max_pool')(X)
    X = Flatten()(X)
    X = Dense(1, activation='sigmoid', name='fc')(X)
    model = Model(inputs = X_input, outputs = X, name='HappyModel')
    
    return model
```

```python
happyModel = HappyModel((64,64,3))
happyModel.compile(optimizer="Adam", loss="binary_crossentropy", metrics=["accuracy"])
happyModel.fit(x=X_train, y=Y_train, epochs=10, batch_size=16)
preds = happyModel.evaluate(x=X_test, y=Y_test)
print()
print ("Loss = " + str(preds[0]))
print ("Test Accuracy = " + str(preds[1]))
```

#### 保存模型

```python
# 保存模型 (architecture + weights + optimizer state)
happyModel.save('advanced/my_model.h5')
```

```python
# 以另外一个变量的方式重新载入模型
model = load_model('advanced/my_model.h5')
```

```python
# 查看模型在 验证集 上的效果，和原始模型在 验证集 上的效果一致
model.evaluate(x=X_test, y=Y_test)
```

#### 保存模型的 architecture

```python
# 保存模型的 architecture
json_string = model.to_json()
with open('advanced/architecture.json','w') as f:
    f.write(json_string)
```

```python
with open('advanced/architecture.json','r') as f:
    json_load = f.read()

from keras.models import model_from_json
model_json = model_from_json(json_load)
```

```python
# 由于仅仅是 load 进来了一个模型的框架，还没训练，因此此时 eveluate 会报错
model_json.evaluate(x=X_test, y=Y_test)
```

#### 保存模型的权重

```python
# 保存模型的权重
model.save_weights('advanced/my_model_weights.h5')
```

```python
new_model = HappyModel((64,64,3))
new_model.load_weights('advanced/my_model_weights.h5')
new_model.compile(optimizer="Adam", loss="binary_crossentropy", metrics=["accuracy"])
```

```python
new_model.evaluate(x=X_test, y=Y_test)
```
